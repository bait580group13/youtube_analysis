## Purpose
Youtube is the most famous video-sharing website in the world. With increasing audience and popularity amongst the masses, it continues to be a vital tool for promotion across various industries. The platform is regarded mainly as a video-sharing site where users can watch and upload videos. However, its other features like comments, likes/dislikes, polls, and community posts also provide interaction between users. The factors which establish Youtube as a potential marketing platform are as follows:

 - Audience reach: over 2 billion monthly active users

 - Diverse customer base: age of users range from 18 years to 56+ years, with 62% male and 38% female users

 - Accessibility: There are more than a hundred local versions of YouTube in over 80 languages

It is these features that drive the objective of our analysis. We want to model and get a deeper insight into the factors that make the content impactful on youtube. The main verticals of our study are:

 - To analyze which features (likes/dislikes/comments etc.) drive “Views” across different categories such as entertainment, politics and sports

 - To investigate which keywords in the comments/descriptions attribute to “View” count

## Results
On basis of our four analyses we see the following results:

Baseline linear regression with target field as "Views"
 - A video disliked has a higher chance of more views when compared to a liked video 
 - A video tagged to "sports" category will have higher views compared to a video tagged as "music" category
 - Comments do not necessarily increase the views. If not, they might have a negative effect

Linear regression Model with sentiment analysis
 - Negative sentiment of the description will increase the view count
 - "Comments" do increase the views count
 - A video tagged to "Comedy" category will have lower views compared to a video tagged to "Movies"

Count analysis for all data
 - Positive words leading to increased "views" are available, entertainment, play, producer, exclusive etc.
 - Negative words leading to increased "views" are episodes, link, website, production, tv etc.

 - This analysis shows the top words for each of the categories

 
 NOTE: the causal relationship between the response and predictor variables is out of scope for this analysis


## Conclusion
 - Disabling ratings and enabling comments could lead to increased views
 - Description sentiment exhibits a negative correlation with views i.e. A negatively phrased video description will lead to less viewers
 - “Entertainment”, “play” and “exclusive” exhibit the greatest correlation with views and so where possible, should be incorporated into video descriptions

## Linear Regression Model with sentiment analysis

````python
# Linear Regression Model with sentiment analysis for descriptions

# preparing new DF for sentiment analysis

text_analysis_df = videos.drop(columns = ['channel_title', 'video_id','publish_time','thumbnail_link'])
text_analysis_df.info()
````

````python
# Dropping rows which does not have descriptions

text_analysis_df = text_analysis_df.dropna(axis = 0)
text_analysis_df.info()
````

````python
text_analysis_df['description'] = text_analysis_df['description'].astype('str')
````

````python
# Preprocessing for titles and descriptions (Removing special characters and stopwords)

from textblob import TextBlob
from nltk.corpus import stopwords
import re
import nltk

nltk.download('stopwords')

en_stops = stopwords.words('english')

cleaned_title = []

for x in text_analysis_df['title'].tolist():
    x = re.sub('[^0-9a-zA-Z \t]+','',x)
    x_list = x.split(' ')
    new_xl = []
    for xc in x_list:
        if xc not in en_stops and 'http' not in xc:
            new_xl.append(xc)
    new_xl = filter(lambda x: x != '', new_xl)
    new_x = ' '.join(new_xl)
    cleaned_title.append(new_x)

cleaned_title


cleaned_desc = []

for d in text_analysis_df['description'].tolist():
    d = re.sub('[^0-9a-zA-Z \t]+','', d)
    d_list = d.split(' ')
    new_dl = []
    for dc in d_list:
        if dc not in en_stops and 'http' not in dc:
            new_dl.append(dc)
    new_dl = filter(lambda nd: nd != '', new_dl)
    new_d = ' '.join(new_dl)
    cleaned_desc.append(new_d)

text_analysis_df['cleaned_title'] = cleaned_title
text_analysis_df['cleaned_desc'] = cleaned_desc

text_analysis_df
````

````python
# finding languages for descriptions

from langdetect import detect

title = []

title_lang = []
for text in text_analysis_df['cleaned_title'].tolist():
    try:
        title_lang.append(detect(text))
    except:
        title_lang.append('none')

        
desc_lang = []
for desc in text_analysis_df['cleaned_desc'].tolist():
    try:
        desc_lang.append(detect(desc))
    except:
        desc_lang.append('none')

text_analysis_df['title_language'] = title_lang
text_analysis_df['desc_language'] = desc_lang

text_analysis_df['desc_language'].value_counts()
````

````python
# choosing descriptions written in English

text_analysis_df_en = text_analysis_df.query("desc_language == 'en'")
````

````python
# Adding sentiment and subjectivity for each english description

desc_pol = []
desc_subj = []
for desc in text_analysis_df_en['cleaned_desc'].tolist():
    desc_pol.append(TextBlob(desc).sentiment[0])
    desc_subj.append(TextBlob(desc).sentiment[1])

text_analysis_df_en['desc_sentiment'] = desc_pol
text_analysis_df_en['desc_subjectivity'] = desc_subj

text_analysis_df_en.columns
````

````python
# Choosing variables for the linear regression model

text_analysis_final_df = text_analysis_df_en[['category_name', 'views', 'likes', 'dislikes', 'comment_count',
       'comments_disabled', 'ratings_disabled', 'video_error_or_removed',
       'desc_sentiment', 'desc_subjectivity']]
````

````python
# Correlation table

text_analysis_final_df.corr()
````

````python
# Spliting X and y

NLP_X = text_analysis_final_df.drop(columns = ['views'])
NLP_y = text_analysis_final_df['views']
````

````python
# Preprocessing for categorical X variable

NLP_X['category_name'] = videos['category_name']
NLP_X = pd.get_dummies(NLP_X)
NLP_X
````

````python
# Fitting Linear Regression Model and Summary for the model

lin_reg = sm.OLS(NLP_y, NLP_X).fit()

lin_reg.summary()
````

## Count Analysis for all data

````python
# Choosing variables for the word count analysis

word_count_df = text_analysis_df_en[['cleaned_desc', 'views']]

# Spliting X and y

word_count_X = word_count_df['cleaned_desc']
word_count_y = word_count_df['views']


# Scaling for y as its values are too high

sds = StandardScaler()

word_count_y = pd.DataFrame(word_count_y, columns = ['views'])
word_count_y = sds.fit_transform(word_count_y)
word_count_y
````

````python
# choosing 100 words for counting

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD

cv = CountVectorizer(max_features = 100, min_df = 3, max_df = 0.8, stop_words='english')

word_count_X = cv.fit_transform(word_count_df['cleaned_desc'])
word_count_X = word_count_X.toarray()
````

````python
# Linear Regression model for counts for 100 words

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(word_count_X, word_count_y)
````

````python
# Coefficients of the model

coef_df = pd.DataFrame({
    'words': cv.get_feature_names(),
    'coefficient': lr.coef_[0]
})

coef_df
````



## The 10 rules
These recommendations are based on [6]

1. Tell a story for an audience: interleave explanatory text with code and results to create a computational narrative.

2. Document the process, not just the results: so that others, and yourself later, will understand your reasoning and choices.

3. Use cell divisions to make steps clear: have eaach cell perform one meaningful and documented step, avoid long cells.

4. Modularize code: use and document functions for repeated operations to make the code more readable and save space.

5. Record dependencies: use pip or Conda package manager, with environment.yml (Conda) or requirements.txt (pip). Please indicate package versions explicitly. Ideally, include a listing of dependencies at the bottom of the notebook. Make sure you test your notebook in an environment created from these dependencies, so that you don't add undocumened dependencies.

6. Use version control: https://github.com/jupyter/nbdime is a diffing tool specifically designed for notebooks; you may find it useful.

7. Build a pipeline: place key variable declarations at the top. Test with different parameters, clean up, and run all cells to prepare for potential non-interactive execution.

8. Share and explain your data: if the original data are too big, inlude a sample. Make sure that the data is accessible. Include data description, and any processing done beforehand. Ideally, point to datasets that have been permanently managed and identified by DOIs.

9. Design your notebooks to be read, run, and explored:  
    9.1. Read: store it in a public code repository with a clear README file and a liberal open source license. Generate HTML/PDF versions of the final notebook.  
    9.2. Run: use Binder to run it in the cloud, and/or create a Docker image.  
    9.3. Explore: use interactive widgets (e.g., ipywidgets, or panel) 
    
10. Advocate for open research: ask your colleagues or friends to try to run your notebooks.


__Additional recommendations, beyond the 10 rules:__
1. Do not include shell commands in cells (unless the notebook will be distributed in Docker container.)
2. Keep cell output if other researchers will use the included data, or clear cell output if they will use their own data.
3. Develop your code in a clean virtual environment, to avoid dependency conflicts and streamline packaging.

To create a virtual environment, use
```
conda create -n <enviroment_name> python=3 jupyter -y
```
or

and then activate it.

## Adding table of contents

Having a table of contents below the title and authors will make your notebook look more professional, and will help users navigate it.

The key steps are:
1. In terminal, install jupyter_contrib_nbextensions
```
pip install --user jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
```

2. Then, enable the Table of Contents extension:
```
jupyter nbextension enable toc2/main
```

You will see a new icon "Table of Contents" added to the row of icons. It will toggle the ToC view. Click the Gear icon on the ToC panel to add a ToC cell to your notebook.

## Creating Docker and Binder for your notebook repository

Sharing just the Jupyter notebook .ipynb file is not sufficient for broad re-usability. We request that Jupyter notebooks are submitted to EarthCube in Binders. The Binder service is a free service that allows anyone to run notebooks from their web browser without installing any software and resolving dependencies. 

Follow these steps:
1. Assemble the Jupyter notebooks , environment files, custom external libraries, datasets, and a README, in a single project directory (and sub-directories off of it, as needed). Use GitHub to organize your project as a GitHub repository.

2. Optionally, test your project with repo2docker. See instructions in [2]. This will create a docker image which can be made available in a Docker repository. 

3. Create Binder. Use mybinder.org to create a URL for your notebook Binder (you will need to enter your GitHub repo URL). You can also add a Launch Binder button directly to your GitHub repo, by include the following in your README.md:
```
launch with myBinder
[![Binder](https://mybinder.org/badge.svg)](https://mybinder.org/v2/gh/<path to your repo>)
```

Specific projects may have their own Binder implementations. In this case, include a project-specific Binder reference so that reviewers can execure your code.

Expect the reviewers to use a browser of their choice to run your code, and avoid execution environments that have any special requirements that reviewers may not have (for example, Google colab requires a Google account, which reviewers may not have).
